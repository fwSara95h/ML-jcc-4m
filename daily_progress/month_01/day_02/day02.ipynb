```python

user = None  # {"role": "admin"} # {"role": "member"}

# will raise error unless `user` has a `role` attribute
# is_admin = (user.get("role") == "admin") and (user is not None) 

# should run without error:
is_admin = (user is not None) and (user.get("role") == "admin")
print(is_admin)
```

### MC2 : The Floating Point Trap

```python
print(0.1 + 0.2 == 0.3, f"[{0.1 + 0.2=}]")  
# False due to floating-point precision issues
```

```python
epsilon = 1e-9 # A Data-Science-safe comparison tolerance 
print(abs((0.1 + 0.2) - 0.3) < epsilon, f'[{(0.1 + 0.2) - 0.3=}]') 
```

```python
print(round(0.1 + 0.2, 1) == 0.3, f'[{round(0.1 + 0.2, 1)=}]')
```

### MC4 : The Ternary Packer

For simple cases, we can use single-line conditional expressions instead of more expansive if-statements

```python
score = 85
status = "Pass" if score >= 70 else "Fail" # single-line 
print(status)
```

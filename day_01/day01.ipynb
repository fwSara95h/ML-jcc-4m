### MC1 : The Identity Swap 

```python
x = 100
y = 200
x , y = y , x # teh one-line swap
print(f"{x=} , {y=}")
```

### MC2 : The Type Auditor 

```python
n_in = input("Enter a number: ")
print("Type of n_in :", ni_type:=type(n_in))

n_float = float(n_in) # cast as float
nf_type = type(n_float)
print(f"{ni_type} {n_in=!r} has been cast to {nf_type} {n_float=!r}")
```

### MC3 : The Precision Banker 

```python
bill = 100 / 3
print(f'Unformatted : {bill}\nFormatted : {bill : .3}')
```

### MC4 : The Modulo Architect 

```python
total = int(input("Enter in a raw number of seconds : "))

# cycle 1
h = total // 3600
remainder = total % 3600

# cycle 2
m = remainder // 60
s = remainder % 60

print(f"{total}s = {h}hrs + {m}min + {s}s")
```

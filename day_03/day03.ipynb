### MC1 : The Infinite Guardian

```python
while True: # run forever unless `break` is executed
    pwd = input("Enter password (or type 'stop' to quit): ")
    if pwd == "stop": break # exit clause
    print("Password received.")
print("[Loop ended]")
```

### MC2 : The Accumulator Pattern

```python
n = int(input("Enter N: "))

total = 0 # initialize accumulator
for i in range(1, n + 1): # add every number from 1 to N
    total += i 

print(f"Total sum from 1 to {n} is: {total}")
```

### MC3 : The Efficient Skipper

```python
for i in range(1, 11):
    if i == 5:
        continue # skip 5
    print(i)
```

### MC4 : The String Walker

```python
word = "DATA"

for ch in word: # string is iterable
    print(ch)
```
